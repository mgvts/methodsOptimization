```python
# from conspects page 9-10
# метод дихотомии
#http://cs.mipt.ru/advanced_python/lessons/lab18.html

import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import sympy as sp
import math


def f(value):
    expr = sp.sympify("x^2 - 2*x + 1")
    return expr.subs(x, value)


def calc_min_iterations():
    return sp.log((b - a - delta) / (2 * eps - delta), 2)


x, y, z = sp.symbols('x y z')
sp.init_printing(use_unicode=True)

a = -10
b = 10
eps = 0.001
# delta must be from (0, 2*eps)
delta = 0.0015
N = math.ceil(calc_min_iterations())
print(f"{N =}")

plt.figure(figsize=(b - a, math.fabs(f(a)) + math.fabs(f(b))))
fig, ax = plt.subplots()  # Создание окна и осей для графика


def create_plot():
    x_plt = np.arange(a, b, 0.1)
    f_plt = [f(t) for t in x_plt]
    ax.plot(x_plt, f_plt)

    plt.yticks()  # можно явно указать, как подписывать ось
    plt.xticks()
    plt.xlabel('ось абсцисс')
    plt.ylabel('ось ординат')
    ax.grid()


def add_point(t, c="red"):
    ax.scatter(t, f(t), c=c)  # отображение точки красным цветом


create_plot()

x1 = (a + b - delta) / 2
x2 = (a + b + delta) / 2
t1 = (b - x1) / (b - a)
t2 = (x2 - a) / (b - a)
#t1 and t2 must be lim = 1/2


for i in range(N):
    print(f"{x1 = } {x2 = } {f(x1) = } {f(x2) = }")
    add_point(x1, c="red")
    add_point(x2, c="blue")

    #1 step
    x1 = (a + b - delta) / 2
    x2 = (a + b + delta) / 2

    #2 step
    if f(x1) <= f(x2):
        b = x2
    else:
        a = x1

    #3 step
    eps_i = (b - a) / 2
    if eps_i <= eps:
        break

#4 step
X = (a + b) / 2
print(X)

plt.show()
```
